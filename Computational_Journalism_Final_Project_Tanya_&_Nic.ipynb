# The Great Supply Chain Disruption

### The bottlenecks in the global supply of products, triggered by lockdowns during the pandemic, were supposed to be over by now. 
### But complications continue.

Delays and disruptions have affected many industries. A prime example is the hearing aid industry. In 2020, **fewer hearing aids were sold** in the United States than during any other year in the last decade. At the worst of times, imports of hearing aids into the U.S. **dropped to a third of the volume** recorded pre-pandemic.

```python
#creating a line chart showing percent change in US hearing aid sales vs all retail sales 

#pulling data, hand-compiled from US Census data about sales and data scraped from Hearing Review, which cited the Hearing Industries Association
url = 'https://raw.githubusercontent.com/nicwalter/hearingaids/main/Change%20in%20Hearing%20Aid%20Sales%2C%20Percentage%20Change%20From%20Previous%20Year%20V3.csv'

import pandas as pd
saleschange = pd.read_csv(url)

import altair as alt

saleschange['year'] = saleschange['year'].apply(str)

alt.Chart(saleschange).mark_line().encode(
    x='year', 
    y='percent change',
    color='product'
).properties(
    width = 600, 
    height = 200
).configure_range(
    category={'scheme': 'yellowgreenblue'}
).properties(
    title = "Percentage Change in Sales, Year Over Year")
```

```python
#graphic the monthly values of hearing aid imports into the u.s. over the last 10 years

url5 = 'https://raw.githubusercontent.com/nicwalter/hearingaids/main/Imports%20of%20Hearing%20Aids%2C%20Monthly%20Total.csv' 

import pandas as pd
monthlyimports = pd.read_csv(url5,parse_dates=["Time"])

monthlyimports['Customs Value (Gen) ($US)'] = monthlyimports['Customs Value (Gen) ($US)'].str.replace(',','')
monthlyimports['Customs Value (Gen) ($US)'] = monthlyimports['Customs Value (Gen) ($US)'].astype('float')

import altair as alt

alt.Chart(monthlyimports).mark_area().encode(
    alt.X('Time:T'),
    alt.Y('Customs Value (Gen) ($US):Q')
).properties(
    title = "Monthly Imports of Hearing Aids Into the U.S."
).properties(
    width = 650, 
    height = 200)
```

It's clear that **suppliers were unable to get their products to the people who needed them**, at least for a short period of time.

Since 2020, hearing aid sales have increased exponentially. A number of factors have fueled this upward trend. 
* Pent up consumer demand for hearing aids.
* The number of people with hearing loss is increasing. 
* Operations experts have been able to smooth out some of the kinks in the supply chain. 

Collectively, this led to more hearing aids being sold in the U.S. in 2021 than any other year on record.

## What were some of the Bottlenecks?

1. Many raw material providers and device manufacturers temporarily closed their 
facilities. This, in turn, slowed (and at some moments, stopped) the assembly of new devices.

2. Ports faced extreme congestion as they struggled to adapt to the unpredictable stop and start of supply across products of all types.

3. In March of 2020, over 70% of hearing aid retailers (i.e., audiology clinics) reported they had reduced their hours. 38% said they were not seeing patients at all.

**Hearing aid shortages are particularly notable given that they affect a vulnerable population.** 

Without assistive technology, those with hearing loss are more susceptible to social and cognitive decline. The hardships of 2020 disproportionately impacted lower income people with hearing loss to obtain hearing aids.

## Consumers

> "Those in low income communities experienced a greater impact because they have fewer resources to overcome the impacts of this situation. # What next?

## Disruptions will continue for awhile

We also interviewed Dr. Willy Shih, an economist specializing in supply chain dynamics and a professor at Harvard Business School. 
He noted that excessive fiscal stimulus could be creating a **buying binge** which is making supply chain issues worse for many products being imported from abroad. 

He anticipates that the current Covid-19 wave in China could cause disruption in June 2022 due to major international ports closing for a short period. 

> "America has been on a buying binge. If you talk to most people in freight and logistics, they say it's this buying binge that's overwhelmed the capacity of ports and logistics networks. Then, because you have so many bottlenecks, other people run short on parts. It all just propagates. There's a big interconnected, tangled mess right now, and anytime you spend that long, making a mess like that, it's gonna take a while to sort out."

## Demand is set to rise.

The number is increasing globally of the number of people who require hearing aids. 

Estimates range in terms of how many people actually experience hearing loss, but it seems to be around **20% of the U.S. population**. The variablity across sources is because of the the percieved social stigma around having hearing loss. 

The graph here shows data from the Global Burden of Disease (GBD) study, a survey led by the Institute for Health Metrics and Evaluation at the University of Washington. 

This survey includes data about human health from 204 countries over the last 30 years. Other analyses done with GBD data predict that the number of people with hearing loss is set to continue to rise until 2050. 

Halloran said: 
> "If you look worldwide, one person out of five uses amplification that could. That means four out of five don't use it. And there's a whole host of reasons why that is. …If there was a capability as an intermediate step, then I think people will start getting benefits from amplification earlier in their journey. And so [the FDA regulation change] will expand the market, not contract the market.”

## Over The Counter Hearing Aids - New FDA Regulation

FDA regulations currently require that hearing aids be sold by a medical professional, via a prescription. Later this year, those regulatations [are expected](https://https://www.hearingloss.org/new-hearing-aid-options-are-coming-in-2022/) to be lifted, in what would be a groundbreaking shift for the industry. Not only would this make purchasing a hearing aid easier for a large number of people, it would fundamentally restructure the anatomy of the international hearing aid supply network.
